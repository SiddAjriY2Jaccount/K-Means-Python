```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random as rd
import seaborn as sns
```

```python
dataset = pd.read_csv('College.csv')
dataset.head()
```

```python
dataset['Enrollment_Rate'] = (dataset['Enroll']/dataset['Accept'])*100.0
dataset.head()
```

```python
dataset['Acceptance Rate'] = (dataset['Accept']/dataset['Apps'])*100.0
dataset.head()
```

```python
dataset.shape
```

```python
######################       KNN      #######################


# First Feature
X = list(dataset['Expend'])

# Second Feature
Y = list(dataset['Enrollment_Rate'])

# Label/Target varible
label = list(dataset['Private'])
```

```python
# Import LabelEncoder
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
label_encoded = list(le.fit_transform(label))

label_encoded
```

```python
features = list(zip(X,Y))

#Combining features
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=8)

# Train the model using the training sets
model.fit(features,label_encoded)
```

```python
#Predict Output - 1 indicates private and 0 indicates Public

predicted = model.predict([[8735,30.628988]]) #Testing for Adrian College, Expend = 8735, Enrollment_Rate = 30.628988, which is a Private college
print(predicted)
```
