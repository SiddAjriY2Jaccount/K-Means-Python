```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random as rd
import seaborn as sns
```

```python
dataset = pd.read_csv('College.csv')
dataset.head()
```

```python
sns.set_style('whitegrid')
sns.lmplot('Room.Board','Grad.Rate',data=dataset, hue='Private',
           palette='coolwarm',height=6,aspect=1,fit_reg=False)
```

```python
X = dataset.iloc[:, [10, 18]].values #Choosing Room.Board and Grad.Rate as Atrributes
X
```

```python
print(X.shape)

m = X.shape[0] #number of training examples
n = X.shape[1] #number of features. Here n will be 2, naturally

print("m,n",m,n)

epochs = 100 #max no. of iterations
```

```python
K = 2  # number of clusters, for now, change it later
```

```python
Centroids = np.array([]).reshape(n,0) #Centroids is a n x K dimensional matrix, where each column will be a centroid for one cluster
```

```python
print(np.array(Centroids))
```

```python
for i in range(K):
    rand = rd.randint(0,m-1)
    Centroids = np.c_[Centroids,X[rand]]

Centroids
```

```python
Output={} #Dict with key as Cluster no. and Values as each data point in the cluster
```

```python
'''
EuclidianDistance = np.array([]).reshape(m,0)
for k in range(K):
    tempDist = np.sum((X-Centroids[:,k])**2,axis=1)
    EuclidianDistance = np.c_[EuclidianDistance,tempDist]
C = np.argmin(EuclidianDistance,axis=1)+1

C
'''
```

```python
for i in range(epochs):
     #step 2.a
      EuclidianDistance=np.array([]).reshape(m,0)
      for k in range(K):
          tempDist=np.sum((X-Centroids[:,k])**2,axis=1)
          EuclidianDistance=np.c_[EuclidianDistance,tempDist]
      C=np.argmin(EuclidianDistance,axis=1)+1
     #step 2.b
      Y={}
      for k in range(K):
          Y[k+1]=np.array([]).reshape(2,0)
      for i in range(m):
          Y[C[i]]=np.c_[Y[C[i]],X[i]]
     
      for k in range(K):
          Y[k+1]=Y[k+1].T
    
      for k in range(K):
          Centroids[:,k]=np.mean(Y[k+1],axis=0)
      Output=Y
    
Output
```

```python
color=['red','blue','green','cyan','magenta']
labels=['cluster-1','cluster-2','cluster-3','cluster-4','cluster-5']
for k in range(K):
    plt.scatter(Output[k+1][:,0],Output[k+1][:,1],c=color[k],label=labels[k])
plt.scatter(Centroids[0,:],Centroids[1,:],s=300,c='yellow',label='Centroids')
plt.xlabel('Room.Board')
plt.ylabel('Grad.Rate')
plt.legend()
plt.show()
```
